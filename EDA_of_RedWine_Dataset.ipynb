✔ Identify patterns, relationships, and trends<br>
✔ Helps you choose appropriate algorithms based on data characteristics<br>

# Visualizing data
it helps you see patterns, relationships, and outliers that numbers alone can't reveal.

Visualization makes it easier to understand distributions, correlations, and trends.

**Common Visualizations in EDA**

* Histograms → Show the distribution of a single feature.

* Box plots → Highlight outliers and spread.

* Scatter plots → Show relationships between two features.

* Correlation heatmaps → Show how features are related to each other.

* Count plots → Show frequency of categories in categorical data.

```python
# Histogram for 'alcohol'
import seaborn as sns
sns.histplot(x=df['alcohol'], bins=10) # histplot divides the entire range of alcohol values into 10 equal-sized intervals (bins).
                                     # For each bin, it counts how many samples fall within that interval.
```

NOTE :<br>
x=  → the data will be plotted along the horizontal axis (X-axis)

y=  → the data will be plotted along the vertical axis (Y-axis)

```python
# Box plot – Check for outliers in 'residual sugar'
import matplotlib.pyplot as plt
sns.boxplot(y=df['residual sugar'])
plt.title("Boxplot of Residual Sugar")
plt.show()
```

```python
# Scatter plot – See how 'alcohol' relates to 'quality'
import matplotlib.pyplot as plt
import seaborn as sns
sns.scatterplot(x=df['alcohol'], y=df['quality'])
plt.title("Alcohol vs Quality")
plt.xlabel("Alcohol")
plt.ylabel("Quality")
plt.show()
```

**Observations from your plot:**<br>
Most wines have quality 5, 6, or 7. As alcohol increases, the quality tends to be slightly higher.<br>
There’s a positive relationship between alcohol and quality—but it’s not perfect. There’s still overlap and there is a variation.<br>
The points don’t form a straight, clean line—they’re scattered, with overlap and variability.

```python
# Correlation heatmaps

import seaborn as sns
sns.heatmap(df.corr())
```

**What is Correlation?**<br>

Correlation measures how two numerical features are related.<br>

How is Correlation Useful in EDA?<br>

✔ Helps identify which features are most related to the target variable<br>
✔ Guides feature selection or engineering<br>

| Correlation | Meaning                                                              |
| ----------- | -------------------------------------------------------------------- |
| +1          | Perfect positive correlation (as one increases, the other increases) |
| 0           | No correlation (no relationship)                                     |
| -1          | Perfect negative correlation (as one increases, the other decreases) |


```python
# To represent values of correlation in the heatmap

import matplotlib.pyplot as plt
plt.figure(figsize=(10,8))        # This sets the size of the heatmap you’re about to draw. 
                                 # figsize=(10, 8) means the width is 10 units and the height is 8 units.
sns.heatmap(df.corr(), annot=True) # annot=True :  This means show the actual numbers inside each cell.
```

<u>**How correlation is calculated?**<br></u>
The most common method used in .corr() is **Pearson’s correlation coefficient**.<br>
$$r = \frac{\sum(x_i - \bar{x})(y_i - \bar{y})}{\sqrt{\sum(x_i - \bar{x})^2 \cdot \sum(y_i - \bar{y})^2}}$$



**Using Spearman Formula.**<br>
$$r_s = 1 - \frac{6\sum d_i^2}{n(n^2-1)}$$

```python
# To represent values of correlation in the heatmap using spearman method

import matplotlib.pyplot as plt
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(method='spearman'), annot=True)
```

** Difference between pearsons and spearmans correlation: **<br>
i)Pearsons measure linear relationship between two features with a constant rate (which is a straight line).
  whereas spearmans measure Monotonic realtionship(as one variable increases, the other either: Always increases, or Always decreases ,But the rate of change doesn’t have to be constant!).

ii)Pearson is calculated based on actual values while spearman is based on ranks.

iii)pearson is used when realtionship is linear and normally distributed while spearman used for ord
